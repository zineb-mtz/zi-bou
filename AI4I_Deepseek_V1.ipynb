```python
# -*- coding: utf-8 -*-
"""AI4Imaging-Hackathon-2024-Cardiac-CNN-Corrected.ipynb"""

# ========================
# INITIAL SETUP
# ========================
print("â³ Starting setup...")

from google.colab import drive
import os
import nibabel as nib
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
from scipy.ndimage import zoom  # Added missing import

print("âœ… Basic imports completed")

# ========================
# DATA PATHS CONFIGURATION
# ========================
print("\nğŸ” Configuring data paths...")

drive.mount('/content/drive')
print("âœ… Google Drive mounted")

# Corrected path to Al4l (lowercase L)
base_dir = '/content/drive/My Drive/'
data_dir = os.path.join(base_dir, 'KAGGLE/Al4l/data')  # Changed AI4l to Al4l
train_dir = os.path.join(data_dir, 'drive-download-20250107T191042Z-001/train')
test_dir = os.path.join(data_dir, 'drive-download-20250107T191042Z-001/test')

print(f"ğŸ“‚ Base directory: {base_dir}")
print(f"ğŸ“ Data directory: {data_dir}")
print(f"ğŸ“‚ Training directory: {train_dir}")

# ========================
# DATA PROCESSING UTILITIES
# ========================
print("\nğŸ”§ Initializing data processing utilities...")

def process_compressed_files(directory, max_iterations=10):
    """Safer compressed file processor with iteration limit"""
    print(f"\nğŸ” Searching for compressed files in: {directory}")

    for iteration in range(max_iterations):
        print(f"\nğŸ”„ Processing iteration {iteration+1}/{max_iterations}")
        compressed_files = []

        for root, _, files in os.walk(directory):
            for file in files:
                if file.endswith(('.gz', '.zip')):
                    compressed_files.append(os.path.join(root, file))

        if not compressed_files:
            print("âœ… No more compressed files found")
            return

        print(f"ğŸ“¦ Found {len(compressed_files)} compressed files to process")

        for i, compressed_file in enumerate(compressed_files, 1):
            print(f"\nğŸ“¦ Processing file {i}/{len(compressed_files)}: {os.path.basename(compressed_file)}")

            try:
                if compressed_file.endswith('.gz'):
                    output_path = compressed_file[:-3]
                    with gzip.open(compressed_file, 'rb') as f_in:
                        with open(output_path, 'wb') as f_out:
                            shutil.copyfileobj(f_in, f_out)
                    os.remove(compressed_file)
                    print(f"âœ… Extracted & removed .gz file")

                elif compressed_file.endswith('.zip'):
                    output_dir = compressed_file[:-4]
                    with zipfile.ZipFile(compressed_file, 'r') as zip_ref:
                        zip_ref.extractall(output_dir)
                    os.remove(compressed_file)
                    print(f"âœ… Extracted & removed .zip file")

            except Exception as e:
                print(f"âŒ Error processing {compressed_file}: {str(e)}")
                continue

    print(f"âš ï¸ Reached maximum iterations ({max_iterations}). Stopping compression processing.")

print("âœ… File processing utilities initialized")

# ========================
# DATA PREPROCESSING
# ========================
print("\nğŸš€ Starting data preprocessing...")

# Get actual patient IDs with progress
print("\nğŸ” Scanning patient directories...")
patient_ids = [pid for pid in os.listdir(train_dir)
              if os.path.isdir(os.path.join(train_dir, pid))]
print(f"âœ… Found {len(patient_ids)} patient directories")

def locate_patient_data(patient_id):
    """Enhanced data locator with progress tracking"""
    print(f"\nğŸ” Processing patient {patient_id}")
    patient_path = os.path.join(train_dir, patient_id)

    if not os.path.exists(patient_path):
        print(f"âŒ Directory not found: {patient_path}")
        return None

    files = os.listdir(patient_path)
    print(f"ğŸ“ Found {len(files)} files in directory")

    required_files = {
        'systolic_frame': None,
        'diastolic_frame': None,
        'systolic_mask': None,
        'diastolic_mask': None
    }

    for f in files:
        if 'frame01' in f and 'gt' not in f and f.endswith('.nii'):
            required_files['systolic_frame'] = f
        elif 'frame01_gt' in f and f.endswith('.nii'):
            required_files['systolic_mask'] = f
        elif 'frame' in f and 'gt' not in f and 'frame01' not in f and f.endswith('.nii'):
            required_files['diastolic_frame'] = f
        elif 'frame' in f and 'gt' in f and 'frame01' not in f and f.endswith('.nii'):
            required_files['diastolic_mask'] = f

    missing = [k for k,v in required_files.items() if v is None]
    if missing:
        print(f"âŒ Missing files: {', '.join(missing)}")
        return None

    print("âœ… All required files found")
    return [
        nib.load(os.path.join(patient_path, required_files['systolic_frame'])),
        nib.load(os.path.join(patient_path, required_files['diastolic_frame'])),
        nib.load(os.path.join(patient_path, required_files['systolic_mask'])),
        nib.load(os.path.join(patient_path, required_files['diastolic_mask']))
    ]

def preprocess_patient(patient_id, target_spacing=(1.0,1.0,1.0), target_shape=(256,256,16)):
    """Full preprocessing pipeline with progress tracking"""
    print(f"\nğŸ”§ Preprocessing {patient_id}")

    try:
        images = locate_patient_data(patient_id)
        if images is None:
            return None

        print("ğŸ”„ Resampling volumes...")
        processed = []
        for i, img in enumerate(images):
            is_label = i >= 2  # Masks are label data
            processed.append(
                zoom(img.get_fdata(),
                     [img.header.get_zooms()[d]/target_spacing[d] for d in range(3)],
                     order=0 if is_label else 3)
            )

        print("ğŸ”„ Cropping/padding volumes...")
        for i in range(len(processed)):
            current_shape = processed[i].shape
            new_array = np.zeros(target_shape)
            slices = tuple([slice(0, min(dim, target_shape[j]))
                          for j, dim in enumerate(current_shape)])
            new_array[slices] = processed[i][slices]
            processed[i] = new_array

        print("ğŸ”„ Combining frames...")
        combined = np.stack((processed[0], processed[1]), axis=-1)

        print("ğŸ“„ Loading label...")
        with open(os.path.join(train_dir, patient_id, 'gt.txt'), 'r') as f:
            label = f.read().strip()

        print(f"âœ… Successfully processed {patient_id}")
        return combined, label

    except Exception as e:
        print(f"âŒ Error processing {patient_id}: {str(e)}")
        return None

# ========================
# MAIN PROCESSING
# ========================
print("\nğŸš€ Starting main processing pipeline...")

X, y = [], []
success_count = 0

for idx, pid in enumerate(patient_ids, 1):
    print(f"\nğŸ“¦ Processing patient {idx}/{len(patient_ids)} ({pid})")
    result = preprocess_patient(pid)

    if result:
        X.append(result[0])
        y.append(result[1])
        success_count += 1
        print(f"âœ… Success rate: {success_count}/{idx} ({success_count/idx:.1%})")
    else:
        print(f"âš ï¸ Skipped patient {pid}")

print(f"\nğŸ“Š Final dataset: {len(X)} samples ({len(y)} labels)")

# Convert to numpy arrays
print("\nğŸ”„ Converting to numpy arrays...")
X = np.array(X)
y = np.array(y)

# Encode labels
print("ğŸ”¢ Encoding labels...")
le = LabelEncoder()
y_encoded = le.fit_transform(y)
y_categorical = tf.keras.utils.to_categorical(y_encoded, num_classes=5)

# Train-validation split
print("âœ‚ï¸ Splitting dataset...")
X_train, X_val, y_train, y_val = train_test_split(
    X, y_categorical,
    test_size=0.2,
    random_state=42,
    stratify=y_encoded
)

# ========================
# MODEL BUILDING
# ========================
print("\nğŸ§  Building 3D CNN model...")

def build_3d_cnn(input_shape=(256,256,16,2), num_classes=5):
    print("ğŸ”¨ Constructing model architecture...")
    model = tf.keras.Sequential([
        tf.keras.layers.InputLayer(input_shape=input_shape),

        # Conv Block 1
        tf.keras.layers.Conv3D(32, (3,3,3), activation='relu', padding='same'),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.MaxPool3D((2,2,2)),
        tf.keras.layers.Dropout(0.2),

        # Conv Block 2
        tf.keras.layers.Conv3D(64, (3,3,3), activation='relu', padding='same'),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.MaxPool3D((2,2,2)),
        tf.keras.layers.Dropout(0.3),

        # Conv Block 3
        tf.keras.layers.Conv3D(128, (3,3,3), activation='relu', padding='same'),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.MaxPool3D((2,2,2)),
        tf.keras.layers.Dropout(0.4),

        # Classification Head
        tf.keras.layers.GlobalAveragePooling3D(),
        tf.keras.layers.Dense(256, activation='relu'),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.Dropout(0.5),
        tf.keras.layers.Dense(num_classes, activation='softmax')
    ])
    print("âœ… Model architecture built")
    return model

model = build_3d_cnn()
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
model.summary()

# ========================
# MODEL TRAINING
# ========================
print("\nğŸ‹ï¸ Starting model training...")

callbacks = [
    tf.keras.callbacks.EarlyStopping(
        patience=10,
        restore_best_weights=True,
        verbose=1
    ),
    tf.keras.callbacks.ReduceLROnPlateau(
        factor=0.5,
        patience=3,
        verbose=1
    )
]

history = model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=50,
    batch_size=4,
    callbacks=callbacks,
    verbose=1
)

print("\nğŸ‰ Training completed!")

# ========================
# VISUALIZATION
# ========================
print("\nğŸ¨ Generating visualizations...")

# Training History
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train')
plt.plot(history.history['val_accuracy'], label='Validation')
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train')
plt.plot(history.history['val_loss'], label='Validation')
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend()
plt.show()

print("âœ… Visualizations generated")
```
